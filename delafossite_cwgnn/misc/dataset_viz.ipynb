```python
import os
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def plot_angles_3d_interactive(file_path):
    """
    Loads CSV file with angles and structure_type columns,
    then plots an interactive 3D scatter plot with Plotly.

    Args:
        file_path (str): Path to the CSV file.
    """
    # Load data
    df = pd.read_csv(file_path)

    # Check for required columns
    required_cols = {"structure_type", "ABC_angle", "BCA_angle", "CAB_angle"}
    if not required_cols.issubset(df.columns):
        missing = required_cols - set(df.columns)
        raise ValueError(f"Missing columns in CSV: {', '.join(missing)}")

    # Create Plotly 3D scatter plot
    fig = px.scatter_3d(
        df,
        x='ABC_angle',
        y='BCA_angle',
        z='CAB_angle',
        color='structure_type',
        hover_name='id',  # Optional, remove if 'id' is not present
        size_max=10,
        title="3D Angle Space by Structure Type"
    )

    fig.update_layout(
        scene=dict(
            xaxis_title='ABC_angle',
            yaxis_title='BCA_angle',
            zaxis_title='CAB_angle'
        )
    )

    fig.show()
```

```python
dataset_path = "/mnt/d/ORNL Internship Summer 2025/Delafossite ML/bigvasp"
dataset_file = "target_structure_type_concepts_triplet_angles.csv"

plot_angles_3d_interactive(os.path.join(dataset_path, dataset_file))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_angles_2d(file_path):
    df = pd.read_csv(file_path)

    sns.scatterplot(data=df, x='ABC_angle', y='BCA_angle', hue='structure_type', s=70)
    plt.xlabel('ABC_angle')
    plt.ylabel('BCA_angle')
    plt.title('ABC vs BCA Angle by Structure Type')
    plt.legend(title='Structure Type')
    plt.show()

plot_angles_2d(os.path.join(dataset_path, dataset_file))
```

```python
import seaborn as sns
import pandas as pd

def plot_pairplot(file_path):
    df = pd.read_csv(file_path)
    sns.pairplot(df, vars=['ABC_angle', 'BCA_angle', 'CAB_angle'], hue='structure_type', diag_kind='hist', height=2.5)

plot_pairplot(os.path.join(dataset_path, dataset_file))
```
